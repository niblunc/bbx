# BBX fMRI Scan Notes

```python
import os, glob
from IPython.core import display as ICD
import pandas as pd
import ipywidgets as widgets

pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
```

```python

s1_notes_path='C:\\Users\\19802\\Documents\\nibl\\bbx_inputs\\bbx_w1scan_raw_11102020.xlsx'
s1_notes=pd.read_excel(s1_notes_path)
s1_notes.drop([0,1], axis=0,inplace=True)
s1_notes.set_index("participantID",inplace=True)

s2_notes_path='C:\\Users\\19802\\Documents\\nibl\\bbx_inputs\\bbx_w2scan_raw_11112020.xlsx'
s2_notes=pd.read_excel(s2_notes_path)
s2_notes.drop([0,1], axis=0,inplace=True)
s2_notes.set_index("participantID",inplace=True)

s1_notes=s1_notes.filter(like='scannotes', axis=1)
s2_notes=s2_notes.filter(like='scannotes', axis=1)
subject_list=list(s1_notes.index.values)
```

```python
def display_notes(subject=subject_list[0]):
    display(s1_notes.loc[subject,:])

# run widget that calls help functions above
notes = widgets.Select(options=subject_list)
widgets.interactive(display_notes, subject=notes)
```
